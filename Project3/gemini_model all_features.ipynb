```python
# Drop 'city' and 'region' columns
featured_data_encoded = featured_data_embeddings_dropped.drop(columns=['city', 'region'])

# One-hot encode the 'country' categorical feature
featured_data_encoded = pd.get_dummies(featured_data_encoded, columns=['country'])

featured_data_encoded.head()
```

```python
# Rename the 3 country columns in featured_data_encoded
country_columns = [
    'country_Amerika Birleşik Devletleri',
    'country_Canada',
    'country_Türkiye',
    'country_United States'
]
new_country_columns = ['country_1', 'country_2', 'country_3', 'country_4']

featured_data_encoded = featured_data_encoded.rename(
    columns=dict(zip(country_columns, new_country_columns))
)

featured_data_encoded.head()
```

```python
import google.generativeai as genai

# Configure API key
genai.configure(api_key="AIzaSyDGjPtWmRl2v7B4EkEi_uTEhEdDcpHEY1Y")

# Prepare the prompt for Gemini model using all features except 'job_title'
feature_columns = [col for col in featured_data_encoded.columns if col != 'job_title']
feature_descriptions = {
    'has_hr': "Indicates if the candidate has HR experience (1) or not (0).",
    'word2vec_max_similarity': "Semantic similarity of the candidate's profile to HR-related terms using Word2Vec to the job title being searched for.",
    'glove_max_similarity': "Semantic similarity using GloVe embeddings to the job title being searched for.",
    'sentence_max_similarity': "Semantic similarity using sentence embeddings to the job title being searched for.",
    'seniority_score': "Seniority level of the candidate (higher is more senior). The more the seniority, the higher the score.",
    'similarity_cluster': "Cluster assignment based on GloVe embeddings similarity.",
    'experience_score': "Score representing the candidate's relevant experience. More experience results in a higher score.",
    'connection_score': "Score based on the candidate's professional connections. More connections result in a higher score.",
    'country_1': "Candidate is from country 1.",
    'country_2': "Candidate is from country 2.",
    'country_3': "Candidate is from country 3.",
    'country_4': "Candidate is from country 4.",
    'id': "Unique identifier for the candidate."
}

prompt = (
    "You are an HR expert. Analyze the following candidates and rank them from most to least suitable for HR positions like 'seek human resource' and 'human resource professional'.\n\n"
    "IMPORTANT: Do NOT write any code. Only provide the final ranking as a simple list.\n\n"
    "Ranking criteria (in order of importance):\n"
    "1. HR experience (has_hr = 1 is much better than has_hr = 0)\n"
    "2. Semantic similarity scores (word2vec_max_similarity, glove_max_similarity, sentence_max_similarity) - higher is better\n"
    "3. Seniority and experience scores - higher is better\n"
    "4. Professional connections - higher is better\n\n"
    "Candidates:\n"
)

# Add the candidate feature rows to the prompt
for i, row in featured_data_encoded.iterrows():
    features_str = ", ".join(f"{col}: {row[col]}" for col in feature_columns if col != 'id')
    prompt += f"Candidate {i}: {features_str}\n"

prompt += (
    "\nAnalyze each candidate and return ONLY the ranking as a Python list of indices.\n"
    "Format: [most_suitable_index, second_most_suitable_index, ..., least_suitable_index]\n"
    "Example: [22, 13, 47, 2, 45, ...]\n\n"
    "Do not include any explanations, code, or other text. Only return the list."
)

try:
    # Initialize Gemini model
    model = genai.GenerativeModel('gemini-1.5-flash')
    
    # Generate content with specific instructions
    response = model.generate_content(
        prompt,
        generation_config=genai.types.GenerationConfig(
            temperature=0.1,  # Lower temperature for more consistent output
            max_output_tokens=1000,  # Limit output length
        )
    )
    
    print("Gemini Ranking Response:")
    print(response.text)
    
    # Try to extract the list if Gemini still returns extra text
    import re
    text = response.text.strip()
    
    # Look for a list pattern in the response
    list_match = re.search(r'\[[\d,\s]+\]', text)
    if list_match:
        ranking_list = list_match.group()
        print(f"\nExtracted ranking list: {ranking_list}")
        
        # Convert to actual Python list
        try:
            ranking = eval(ranking_list)
            print(f"Final ranking: {ranking}")
        except:
            print("Could not convert to Python list")
    else:
        print("No list pattern found in response")
        
except Exception as e:
    print(f"Error generating content: {e}")
```

```python
import re

# Extract the ranking list from response.text
ranking_str = re.search(r"\[(.*?)\]", response.text, re.DOTALL).group(1)
ranking_list = [int(x.strip()) for x in ranking_str.split(",") if x.strip().isdigit()]
print("Ranking list:", ranking_list)

# Add a 'gemini_rank' column to featured_data
featured_data['gemini_rank_all_features'] = 0
for rank, idx in enumerate(ranking_list):
    featured_data.at[idx, 'gemini_rank_all_features'] = rank + 1  

# Sort the dataframe by 'gemini_rank'
featured_data_sorted = featured_data.sort_values('gemini_rank_all_features').reset_index(drop=True)
featured_data_sorted.head()
```
