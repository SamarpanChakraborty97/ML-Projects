#### Will be using only glove similarity here.

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

class InitialRanker:
    def __init__(self):
        self.feature_weights = {
            'glove_max_similarity': 0.4,    
            'has_hr': 0.25,
            'seniority_score': 0.15,
            'connection_score': 0.075,
            'country_1': 0.025,
            'country_2': 0.025,
            'country_3': 0.025,
            'country_4': 0.025,
            'experience_score':0.025,
        }
        self.scaler = StandardScaler()
    
    def rank_results(self, features_df):
        # Normalize features
        normalized_features = self.scaler.fit_transform(features_df)
        
        # Calculate weighted score
        scores = np.zeros(len(features_df))
        for i, (feature, weight) in enumerate(self.feature_weights.items()):
            scores += weight * normalized_features[:, i]
        
        # Min-max scale scores to [0, 1]
        min_score = scores.min()
        max_score = scores.max()
        if max_score > min_score:
            scores_scaled = (scores - min_score) / (max_score - min_score)
        else:
            scores_scaled = np.zeros_like(scores)
        
        # Return ranked indices and scaled scores
        ranked_indices = np.argsort(scores_scaled)[::-1]
        return ranked_indices, scores_scaled
```

#### Using a learning to rank model to re-rank the candidates based on the starred candidate

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from torch.distributions import Categorical
import random
```

```python
class LearningToRankModel(nn.Module):
    def __init__(self, feature_dim=8, hidden_dim=128):
        super().__init__()
        self.fc = nn.Sequential(
            nn.Linear(feature_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 1)
        )
    
    def forward(self, features):
        return self.fc(features).squeeze(-1)
```

```python
class LTRReranker:
    def __init__(self, feature_dim=8, lr=0.001):
        self.model = LearningToRankModel(feature_dim)
        self.optimizer = optim.Adam(self.model.parameters(), lr=lr)
        self.criterion = nn.MarginRankingLoss(margin=1.0)
    
    def create_pairwise_data(self, features, starred_idx):
        """Create pairwise training data"""
        pairs = []
        labels = []
        
        starred_features = features[starred_idx]
        
        for i in range(len(features)):
            if i != starred_idx:
                pairs.append((starred_features, features[i]))
                labels.append(1)  # starred candidate should rank higher
        
        return pairs, labels
    
    def train(self, features, starred_idx, epochs=100):
        """Train the ranking model"""
        pairs, labels = self.create_pairwise_data(features, starred_idx)
        
        for epoch in range(epochs):
            total_loss = 0
            
            for (feat1, feat2), label in zip(pairs, labels):
                self.optimizer.zero_grad()
                
                score1 = self.model(feat1.unsqueeze(0))
                score2 = self.model(feat2.unsqueeze(0))
                
                target = torch.tensor([float(label)])
                loss = self.criterion(score1, score2, target)
                
                loss.backward()
                self.optimizer.step()
                total_loss += loss.item()
            
            if epoch % 20 == 0:
                print(f"Epoch {epoch}, Loss: {total_loss/len(pairs):.4f}")
    
    def rerank(self, features):
        """Rerank all candidates based on learned model"""
        with torch.no_grad():
            scores = self.model(features)
            ranked_indices = torch.argsort(scores, descending=True)
            return ranked_indices.numpy()
```

```python
import torch

# Prepare features as a torch tensor for the LTRReranker

# Convert boolean columns to float for PyTorch compatibility
features_tensor = torch.tensor(features_df.astype(float).values, dtype=torch.float32)

# Instantiate the reranker
reranker = LTRReranker(feature_dim=features_tensor.shape[1], lr=0.001)

# Train the reranker using the starred candidate
reranker.train(features_tensor, starred_idx=random_star_idx, epochs=100)

# Get the new ranking
ltr_ranked_indices = reranker.rerank(features_tensor)

# Display the top 10 re-ranked candidates
featured_data_encoded.iloc[ltr_ranked_indices].head(10)
```
